# Iris

## Starter code for basic level - *Colomb-ia*

```python
import pandas as pd
from pandas.plotting import scatter_matrix
import matplotlib.pyplot as plt
import numpy as np
import time
from sklearn import tree
```

### Read dataset

```python
train = pd.read_csv("./data/iris_train.csv", index_col=0)
test = pd.read_csv("./data/iris_test.csv", index_col=0)

trainClass = train['Class']
trainFeatures = train[['Sepal length','Sepal width','Petal length','Petal width']]

testClass = test['Class']
testFeatures = test[['Sepal length','Sepal width','Petal length','Petal width']]
```

### Data Visualization

* Box and whisker plot

```python
plt.rcParams["figure.figsize"] = [16, 9]
train.plot(kind='box', subplots=True, layout=(2,2), sharex=False, sharey=False)
plt.show()
```

* Histogram

```python
train.hist()
plt.show()
```

* Scatter matrix plot

```python
scatter_matrix(train)
plt.show()
```

## Decision Tree

### Training

```python
start_time = time.clock()
clf = tree.DecisionTreeClassifier()
clf = clf.fit(trainFeatures, trainClass)
print(time.clock() - start_time, "seconds")
```

### Testing

```python
accuracy = clf.score(testFeatures, testClass)
print("Accuracy is", accuracy)
```

## SVM

### Training

```python
from sklearn import svm

start_time = time.clock()
clf = svm.SVC()
clf.fit(testFeatures, testClass)
print(time.clock() - start_time, "seconds")
```

### Testing

```python
accuracy = clf.score(testFeatures, testClass)
print("Accuracy is", accuracy)
```
